# **6. Data Preprocessing**

```python
X_train = df_train.drop('label', axis=1) / 255.0
y_train = df_train['label']

X_test = df_test.drop('label', axis=1) / 255.0
y_test = df_test['label']
```

# **7. Visualizing**

```python
y_train.value_counts().sort_index().plot(kind='bar', figsize=(10, 5))
plt.title("Distribution")
plt.xlabel("Label")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
letter = 18
samples = X_train[y_train == letter].iloc[:5]

plt.figure(figsize=(15, 3))
for i in range(5):
    plt.subplot(1, 5, i + 1)
    plt.imshow(samples.iloc[i].values.reshape(28, 28), cmap='gray')
    plt.title(f"Label: {letter}")
    plt.axis('off')
plt.show()
```

# **8. KNN Model**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
KNeighborsClassifier?
```

```python
knn = KNeighborsClassifier()
```

# **9. GridSearchCV**

```python
param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
}
```

```python
grid = GridSearchCV(knn,param_grid,cv=5,scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
print("Best K:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```

```python
best_knn = grid.best_estimator_
```

```python
y_pred = best_knn.predict(X_test)
```

# **10. Evaluation**

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
acc = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {acc*100:.2f}%")
```

```python
cr = classification_report(y_test, y_pred)
print("Classification Report:\n", cr)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(12, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# **11. Displaying Wrong Predictions**

```python
wrong = np.where(y_test != y_pred)[0]

plt.figure(figsize=(15, 4))

for i in range(5):
    idx = wrong[i]
    img = X_test.iloc[idx].values.reshape(28, 28)
    true_label = y_test.iloc[idx]
    predicted_label = y_pred[idx]

    plt.subplot(1, 5, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(f"True:{true_label} | Predicted:{predicted_label}", fontsize=10)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# **12.  PCA for Dimensionality Reduction**

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=50)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
param_grid_pca = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}
```

```python
grid_pca = GridSearchCV(KNeighborsClassifier(),param_grid_pca,cv=5,scoring='accuracy',verbose=2,n_jobs=-1)

grid_pca.fit(X_train_pca, y_train)
```

```python
best_knn_pca = grid_pca.best_estimator_
```

```python
y_pred_pca = best_knn_pca.predict(X_test_pca)
```

# **13. PCA Evaluation**

```python
acc_pca = accuracy_score(y_test, y_pred_pca)
print(f"Test Accuracy after PCA + GridSearch: {acc_pca * 100:.2f}%")
```

```python
cr_pca = classification_report(y_test, y_pred_pca)
print("Classification Report after PCA + GridSearch:\n", cr_pca)
```

```python
cm_pca = confusion_matrix(y_test, y_pred_pca)

plt.figure(figsize=(12, 8))
sns.heatmap(cm_pca, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix after PCA")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# **14. PCA vs Original Accuracy Comparison**

```python
plt.bar(['Before PCA', 'After PCA'], [acc*100, acc_pca*100], color=['tomato', 'skyblue'])
plt.ylabel("Accuracy (%)")
plt.title("KNN Accuracy Comparison")
plt.ylim(50, 100)
plt.show()
```

# **16. Gradio Interface for Prediction**

```python
import gradio as gr
from PIL import Image

with open("knn_model.pkl", "rb") as f:
    model = pickle.load(f)

def predict_sign_language(image):
    if image is None:
        return "Please upload an image first."

    image = image.convert('L')
    image = image.resize((28, 28))
    image_array = np.array(image).reshape(1, -1)
    prediction = model.predict(image_array)[0]
    return f"Predicted Letter: {prediction}"

interface = gr.Interface(
    fn=predict_sign_language,
    inputs=gr.Image(type="pil"),
    outputs="text",
    title="Sign Language Classification with KNN",
    description="Upload a hand sign image to predict the corresponding letter."
)

interface.launch(debug=True)
```
